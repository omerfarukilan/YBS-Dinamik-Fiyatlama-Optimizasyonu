```python
import pandas as pd
import yfinance as yf
import warnings

warnings.filterwarnings("ignore")

# =====================================================
# 1. SATIŞ VERİSİNİ OKUMA
# =====================================================
print("1. Satış verisi okunuyor...")

df_sales = pd.read_csv(
    "online_retail.csv",
    encoding="unicode_escape"
)

# Hatalı kayıtları temizle
df_sales = df_sales[
    (df_sales["Quantity"] > 0) &
    (df_sales["UnitPrice"] > 0)
]

# Tarih formatı
df_sales["InvoiceDate"] = pd.to_datetime(
    df_sales["InvoiceDate"]
).dt.date

print("Satış verisi yüklendi.")

# =====================================================
# 2. DOLAR KURLARINI ÇEKME
# =====================================================
print("2. Dolar kurları indiriliyor...")

start_date = str(df_sales["InvoiceDate"].min())

end_date = (
    pd.to_datetime(df_sales["InvoiceDate"].max())
    + pd.Timedelta(days=1)
).strftime("%Y-%m-%d")

df_fx = yf.download(
    "USDTRY=X",
    start=start_date,
    end=end_date,
    progress=False,
    auto_adjust=False
)

df_fx = df_fx.reset_index()

# MultiIndex varsa düzleştir
if isinstance(df_fx.columns, pd.MultiIndex):
    df_fx.columns = [col[0] for col in df_fx.columns]

print(df_fx.columns)

df_fx = df_fx[['Date', 'Close']]

df_fx.rename(
    columns={
        'Date': 'Tarih',
        'Close': 'Dolar_Kuru'
    },
    inplace=True
)

df_fx['Tarih'] = pd.to_datetime(df_fx['Tarih']).dt.date

print("Dolar kurları indirildi.")

# =====================================================
# 3. TÜKETİCİ GÜVEN ENDEKSİ
# =====================================================
print("3. Güven endeksi okunuyor...")

df_macro = pd.read_excel("evds_veri.xlsx")

print("Sütunlar:")
print(df_macro.columns.tolist())

tarih_sutunu = df_macro.columns[0]
endeks_sutunu = df_macro.columns[1]

df_macro.rename(
    columns={
        endeks_sutunu: "Guven_Endeksi"
    },
    inplace=True
)

df_macro["Yil_Ay"] = pd.to_datetime(
    df_macro[tarih_sutunu],
    errors="coerce"
).dt.to_period("M")

df_macro = df_macro.dropna(
    subset=["Yil_Ay"]
)

# =====================================================
# 4. SATIŞ VERİSİNE AY BİLGİSİ EKLE
# =====================================================
df_sales["Yil_Ay"] = pd.to_datetime(
    df_sales["InvoiceDate"]
).dt.to_period("M")

# =====================================================
# 5. SATIŞ + DOLAR KURU
# =====================================================
print("4. Dolar kuru birleştiriliyor...")

df_merged = pd.merge(
    df_sales,
    df_fx,
    left_on="InvoiceDate",
    right_on="Tarih",
    how="left"
)

df_merged["Dolar_Kuru"] = df_merged["Dolar_Kuru"].ffill()


# =====================================================
# 6. GÜVEN ENDEKSİ BİRLEŞTİRME
# =====================================================
print("5. Güven endeksi ekleniyor...")

df_merged = pd.merge(
    df_merged,
    df_macro[["Yil_Ay", "Guven_Endeksi"]],
    on="Yil_Ay",
    how="left"
)

# =====================================================
# 7. GEREKSİZ SÜTUNLARI SİL
# =====================================================
df_merged.drop(
    columns=["Tarih", "Yil_Ay"],
    inplace=True,
    errors="ignore"
)

# =====================================================
# 8. SONUCU KAYDET
# =====================================================
df_merged.to_excel(
    "birlestirilmis_veri.xlsx",
    index=False
)

print("\nİŞLEM TAMAMLANDI")
print(df_merged.head())

print("\nDosya oluşturuldu:")
print("birlestirilmis_veri.xlsx")
```

```python
import numpy as np

# İşlemlerden önce verinin tarihe göre sıralı olduğundan emin olalım
df_merged = df_merged.sort_values(by=['InvoiceDate', 'StockCode']).reset_index(drop=True)

# ---------------------------------------------------------
# 1. MAKRO RİSK BAYRAĞI
# ---------------------------------------------------------
print("1. Makro Risk Bayrağı oluşturuluyor...")
# Güven endeksi için bir eşik belirliyoruz (Örn: 90'ın altı riskli)
esik_deger = 90
df_merged['Makro_Risk_Bayragi'] = (df_merged['Guven_Endeksi'] < esik_deger).astype(int)

# ---------------------------------------------------------
# 2. KUR VOLATİLİTESİ (Son 7 Günün Standart Sapması)
# ---------------------------------------------------------
print("2. Kur Volatilitesi hesaplanıyor...")
# Kur her satırda aynı gün için tekrar ettiğinden, önce gün bazında tekil bir tablo alıyoruz
df_kur = df_merged[['InvoiceDate', 'Dolar_Kuru']].drop_duplicates().sort_values('InvoiceDate')

# 7 günlük rolling (kayan pencere) ile standart sapma hesapla
df_kur['Kur_Volatilitesi'] = df_kur['Dolar_Kuru'].rolling(window=7, min_periods=1).std()

# İlk gündeki NaN değerini 0 ile doldur (çünkü sapma yok)
df_kur['Kur_Volatilitesi'] = df_kur['Kur_Volatilitesi'].fillna(0)

# Ana tabloya (df_merged) geri birleştir
df_merged = pd.merge(df_merged, df_kur[['InvoiceDate', 'Kur_Volatilitesi']], on='InvoiceDate', how='left')

# ---------------------------------------------------------
# 3. ARZ-TALEP ESNEKLİĞİ
# ---------------------------------------------------------
print("3. Arz-Talep Esnekliği hesaplanıyor...")
# Formül: (Miktar Değişimi / Fiyat Değişimi)
# Sağlıklı olması için her bir ürünü kendi içinde (StockCode) sıralayarak yüzdesel değişimlerine bakıyoruz.

df_merged['Miktar_Degisimi'] = df_merged.groupby('StockCode')['Quantity'].pct_change()
df_merged['Fiyat_Degisimi'] = df_merged.groupby('StockCode')['UnitPrice'].pct_change()

# Fiyatın değişmediği durumlarda (Fiyat_Degisimi = 0) sıfıra bölme hatasını (Sonsuzluk - inf) engellemek için 0'ları NaN yapıyoruz
df_merged['Fiyat_Degisimi'] = df_merged['Fiyat_Degisimi'].replace(0, np.nan)

# Esnekliği Hesapla
df_merged['Esneklik'] = df_merged['Miktar_Degisimi'] / df_merged['Fiyat_Degisimi']

# Sonsuz değerleri ve hesaplanamayan (NaN) ilk satırları temizle
df_merged['Esneklik'] = df_merged['Esneklik'].replace([np.inf, -np.inf], np.nan).fillna(0)

# Gereksiz ara hesaplama sütunlarını düşür
df_merged.drop(columns=['Miktar_Degisimi', 'Fiyat_Degisimi'], inplace=True)

print("Özellik Mühendisliği Tamamlandı!")
print(df_merged[['InvoiceDate', 'StockCode', 'Makro_Risk_Bayragi', 'Kur_Volatilitesi', 'Esneklik']].head(10))
```

```python
import numpy as np

# 1. Simülasyon için elimizdeki en güncel haftanın verilerini alıyoruz
son_hafta_tarihi = df_weekly['Hafta_Baslangici'].max()
df_sim = df_weekly[df_weekly['Hafta_Baslangici'] == son_hafta_tarihi].copy()

# ---> YENİ EKLENEN SATIR: UÇUK ESNEKLİK DEĞERLERİNİ KIRPMA (CLIPPING) <---
# Esnekliği maksimum 0, minimum -5.0 olacak şekilde sınırlandırıyoruz
df_sim['Esneklik'] = df_sim['Esneklik'].clip(lower=-5.0, upper=0.0)

# 2. SENARYO: %5 KUR ŞOKU
kur_soku_orani = 0.05

# 3. KATEGORİZASYON VE HURWICZ ALPHA (α) ATAMASI
# Kural: Fiyat esnekliği -1.0'den büyük (sıfıra yakın) olanlar "Zorunlu İhtiyaç" (Katı)
# Fiyat esnekliği -1.0'den küçük olanlar "İsteğe Bağlı/Lüks" (Esnek) olarak sınıflandırılır.
df_sim['Kategori'] = np.where(df_sim['Esneklik'] >= -1.0, 'Zorunlu İhtiyac', 'Istege Bagli/Luks')

# Hurwicz Kriteri: Zorunlu ürünlerde agresif (α=0.6), Lüks ürünlerde temkinli (α=0.4)
df_sim['Alpha'] = np.where(df_sim['Kategori'] == 'Zorunlu İhtiyac', 0.6, 0.4)

# 4. ZAM ORANI BELİRLEME
df_sim['Uygulanan_Zam_Orani'] = kur_soku_orani * df_sim['Alpha']

# 5. YENİ FİYAT VE TALEP (MİKTAR) GÜNCELLEMESİ
df_sim['Yeni_Fiyat'] = df_sim['UnitPrice'] * (1 + df_sim['Uygulanan_Zam_Orani'])
df_sim['Yeni_Miktar'] = df_sim['Quantity'] * (1 + (df_sim['Uygulanan_Zam_Orani'] * df_sim['Esneklik']))

# Talep 0'ın altına düşemez
df_sim['Yeni_Miktar'] = df_sim['Yeni_Miktar'].apply(lambda x: max(x, 0))

# 6. FİNANSAL ETKİ (ROI) HESAPLAMASI
df_sim['Eski_Ciro'] = df_sim['UnitPrice'] * df_sim['Quantity']
df_sim['Yeni_Ciro'] = df_sim['Yeni_Fiyat'] * df_sim['Yeni_Miktar']
df_sim['Net_Kar_Degisimi'] = df_sim['Yeni_Ciro'] - df_sim['Eski_Ciro']

# SONUÇLARI YAZDIRMA
print("="*50)
print("🎯 FİNANSAL SİMÜLASYON VE HURWICZ SONUÇLARI (DÜZELTİLMİŞ) 🎯")
print("="*50)
print(f"Senaryo: %{kur_soku_orani*100} Kur Şoku")
print(f"Toplam Eski Haftalık Ciro : {df_sim['Eski_Ciro'].sum():,.2f} TL")
print(f"Beklenen Yeni Ciro        : {df_sim['Yeni_Ciro'].sum():,.2f} TL")
print(f"Toplam Net Etki (ROI)     : {df_sim['Net_Kar_Degisimi'].sum():,.2f} TL\n")

print("🏆 En Çok Kâr Getiren 5 Ürün (Strateji Başarısı):")
etki_df = df_sim[['StockCode', 'Kategori', 'Esneklik', 'Uygulanan_Zam_Orani', 'Eski_Ciro', 'Net_Kar_Degisimi']]
etki_df['Uygulanan_Zam_Orani'] = (etki_df['Uygulanan_Zam_Orani'] * 100).round(1).astype(str) + '%'
print(etki_df.sort_values(by='Net_Kar_Degisimi', ascending=False).head())
```
